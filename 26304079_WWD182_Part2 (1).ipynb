**TASK 6.2: Value-Based Segmentation**

#### Value-Based Segmentation Strategy
Customers were segmented based on their **account balance**, which reflects their financial value to the bank.


```python
# Create value-based segments using Balance
df['Value_Segment'] = pd.cut(
    df['Balance'],
    bins=[-1, 50000, 150000, df['Balance'].max()],
    labels=['Low Value', 'Medium Value', 'High Value']
)
```

Customers were categorised based on their **account balance**, which reflects their financial value to the bank.

### Segments Defined

- Low Value  
- Medium Value  
- High Value  

These segments allow the bank to identify customers with different financial importance.

```python
df[['Balance', 'Value_Segment']].head()
```

The number of customers in each value segment was calculated to understand how customers are distributed based on their financial value.

```python
df['Value_Segment'].value_counts()
```

The majority of customers fall within the medium value segment, indicating that most customers maintain moderate account balances.  
A significant portion of customers are in the low value segment, suggesting lower financial engagement.  
The high value segment is the smallest, representing customers with the highest financial value to the bank.
